```python
#@title Download from magnet link
import pandas as pd
import openpyxl
import libtorrent as lt
import os
import shutil
import tempfile
import time
from concurrent.futures import ThreadPoolExecutor, as_completed

# Thng s
file_list_path = '/content/VrPornCensored/Link/OneDriveShareLinkToExcel.xlsx'
sheet_name = 'DownLoadMagnetLink'
table_name = 'DownLoadMagnetLink'
max_workers = 2  # t s lng cng vic ti a

# M file Excel
wb = openpyxl.load_workbook(file_list_path, data_only=True)
ws = wb[sheet_name]

# Tm bng theo tn
if table_name not in ws.tables:
    raise ValueError(f"Table {table_name} not found in the sheet {sheet_name}.")

table = ws.tables[table_name]

# c d liu t bng
data = ws[table.ref]
rows = list(data)
columns = [cell.value for cell in rows[0]]
data = [[cell.value for cell in row] for row in rows[1:]]

df = pd.DataFrame(data, columns=columns)

# Kim tra ct 'Downloaded' v 'Seeded'
if 'Downloaded' not in df.columns or 'Seeded' not in df.columns:
    raise ValueError("Columns 'Downloaded' or 'Seeded' not found in the DataFrame.")

# Lc cc dng cn ti xung
download_df = df[~df['Downloaded'].str.contains('downloaded', case=False, na=False)]

# Lc cc dng cn seed
seed_df = df[df['Downloaded'].str.contains('downloaded', case=False, na=False) & ~df['Seeded'].str.contains('seeded', case=False, na=False)]

# Hm ti file bng magnet link
def download_magnet(magnet_link, save_path):
    print(f"Bt u ti magnet link: {magnet_link}")
    ses = lt.session()
    settings = {
        'listen_interfaces': '0.0.0.0:6881',
        'dht_bootstrap_nodes': 'router.bittorrent.com:6881'
    }
    ses.apply_settings(settings)

    # Ti u ha ci t phin
    session_settings = {
        'announce_to_all_tiers': True,
        'announce_to_all_trackers': True,
        'enable_outgoing_utp': True,
        'enable_incoming_utp': True,
        'enable_outgoing_tcp': True,
        'enable_incoming_tcp': True,
    }
    ses.apply_settings(session_settings)

    params = lt.add_torrent_params()
    params.save_path = save_path
    params.url = magnet_link
    params.flags |= lt.torrent_flags.sequential_download
    torrent_handle = ses.add_torrent(params)

    while not torrent_handle.status().has_metadata:
        time.sleep(1)

    while torrent_handle.status().state != lt.torrent_status.seeding:
        s = torrent_handle.status()
        progress = s.progress * 100
        print(f"\rang ti {s.name}: {progress:.2f}% hon thnh (down: {s.download_rate / 1000:.1f} kb/s up: {s.upload_rate / 1000:.1f} kB/s peers: {s.num_peers})", end='')
        time.sleep(1)

    print(f"\nHon thnh ti {torrent_handle.name()}")
    return {
        'name': torrent_handle.name(),
        'progress': progress,
        'download_rate': s.download_rate / 1000,
        'upload_rate': s.upload_rate / 1000,
        'num_peers': s.num_peers
    }

# Hm seed file n ratio 2.0
def seed_magnet(magnet_link, save_path):
    print(f"Bt u seed magnet link: {magnet_link}")
    ses = lt.session()
    settings = {
        'listen_interfaces': '0.0.0.0:6881',
        'dht_bootstrap_nodes': 'router.bittorrent.com:6881'
    }
    ses.apply_settings(settings)

    params = lt.add_torrent_params()
    params.save_path = save_path
    params.url = magnet_link
    params.flags |= lt.torrent_flags.upload_mode
    torrent_handle = ses.add_torrent(params)

    while not torrent_handle.status().has_metadata:
        time.sleep(1)

    while torrent_handle.status().ratio < 2.0:
        s = torrent_handle.status()
        print(f"\rSeeding {s.name}: ratio {s.ratio:.2f} (up: {s.upload_rate / 1000:.1f} kB/s peers: {s.num_peers})", end='')
        time.sleep(1)

    print(f"\nHon thnh seed {torrent_handle.name()}")
    return {
        'name': torrent_handle.name(),
        'ratio': s.ratio,
        'upload_rate': s.upload_rate / 1000,
        'num_peers': s.num_peers
    }

# Ti v seed file
def process_download(row):
    print(f"Bt u ti xung: {row['Name']}")
    file_name = row['Name']
    magnet_link = row['MagnetLink']
    save_path = row['Save Path']
    result = download_magnet(magnet_link, save_path)
    print(f"Hon thnh ti xung: {result['name']} - {result['progress']:.2f}% hon thnh (down: {result['download_rate']:.1f} kb/s up: {result['upload_rate']:.1f} kB/s peers: {result['num_peers']})")
    if result:
        df.at[row.name, 'Downloaded'] = 'downloaded'


def process_seed(row):
    print(f"Bt u seed: {row['Name']}")
    file_name = row['Name']
    magnet_link = row['MagnetLink']
    save_path = row['Save Path']
    result = seed_magnet(magnet_link, save_path)
    print(f"Hon thnh seed: {result['name']} - ratio {result['ratio']:.2f} (up: {result['upload_rate']:.1f} kB/s peers: {result['num_peers']})")
    if result:
        df.at[row.name, 'Seeded'] = 'seeded'


# Kim tra d liu trong download_df
print("S lng dng cn ti xung:", len(download_df))
print(download_df.head())

# S dng ThreadPoolExecutor  ti v seed ng thi
with ThreadPoolExecutor(max_workers=max_workers) as executor:
    download_futures = {executor.submit(process_download, row): row for _, row in download_df.iterrows()}
    seed_futures = {executor.submit(process_seed, row): row for _, row in seed_df.iterrows()}

    while download_futures or seed_futures:
        for future in as_completed(download_futures):
            try:
                future.result()  # i cho n khi mt link ti xong
                row = download_futures.pop(future)
                print(f"Hon thnh ti xung: {row['Name']}")
                # Kim tra v bt u ti link tip theo nu c
                if not download_df.empty:
                    next_row = download_df.iloc[0]
                    download_df = download_df.iloc[1:]
                    download_futures[executor.submit(process_download, next_row)] = next_row
            except Exception as e:
                print(f"Li khi ti xung: {e}")

        for future in as_completed(seed_futures):
            try:
                future.result()  # i cho n khi mt link seed xong
                row = seed_futures.pop(future)
                print(f"Hon thnh seed: {row['Name']}")
                # Kim tra v bt u seed link tip theo nu c
                if not seed_df.empty:
                    next_row = seed_df.iloc[0]
                    seed_df = seed_df.iloc[1:]
                    seed_futures[executor.submit(process_seed, next_row)] = next_row
            except Exception as e:
                print(f"Li khi seed: {e}")

# Lu li file Excel vi cp nht mi
wb = openpyxl.load_workbook(file_list_path)
ws = wb[sheet_name]

for index, row in df.iterrows():
    ws.cell(row=index + 2, column=columns.index('Downloaded') + 1, value=row['Downloaded'])
    ws.cell(row=index + 2, column=columns.index('Seeded') + 1, value=row['Seeded'])

# S dng tp tm thi  lu workbook
with tempfile.NamedTemporaryFile(delete=False) as tmp:
    temp_file_path = tmp.name
    wb.save(temp_file_path)

# Di chuyn tp tm thi n v tr ch
shutil.move(temp_file_path, file_list_path)
```
